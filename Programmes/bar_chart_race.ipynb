```python
df = pd.read_csv("C:\\1_Programmes Python\\Statistiques\\Classement Elo foot\\New version\\Data Visualization\\log_points.csv", index_col=0)  # index = dates, colonnes = toutes les équipes
df = df.astype(float)
df.index = pd.to_datetime(df.index)
```

Compter 9 frames par seconde de run.

```python
def race(k, duration, spp,bars,title,file_to_save): # k = step, duration en ms, spp = steps per period
    df_ = df.iloc[::k]

    n = len(df_)

    total_frames = n * 10 # steps_per_period = 10

    period_duration = duration/n

    fps = 1000*spp/period_duration # frames per second

    run_duration_min = int(total_frames/(9*60))

    run_duration_sec = int((total_frames/(9*60) - run_duration_min)*60)

    print("duration : ", total_frames/fps) # durée de la vidéo en secondes
    print("total_frames : ", total_frames)
    print("fps : ", fps)
    print("échantillons : ", n)
    print("period_duration : ", period_duration)
    print("durée du run environ : ", run_duration_min, "min", run_duration_sec, "s")

    file = "C:\\1_Programmes Python\\Statistiques\\Classement Elo foot\\New version\\Bart Chart Race\\" + file_to_save + ".mp4"

    colors = [club_colors.get(col, '#aaaaaa') for col in df.columns]

    bcr.bar_chart_race(
        df=df_,
        filename=file,
        orientation='h',
        sort='desc',
        writer="ffmpeg",
        dpi = 144,
        n_bars=bars,
        fixed_order=False,
        steps_per_period=spp, # si interpolate est à False, ça ne sert à rien
        period_fmt = '%Y-%m-%d', # résout le pb d'affichage des dates !
        interpolate_period=True, 
        period_length=period_duration, # en ms
        figsize=(12,6),
        title=title,
        bar_kwargs={'alpha': .7},   # pas de 'color' ici !
        cmap=colors,                 # <-- liste de couleurs par colonne
        filter_column_colors=False,
        title_size=16,
        tick_label_size=10, 
        logos_path="C:\\1_Programmes Python\\Statistiques\\Classement Elo foot\\New version\\Logos\\Logos_resized\\", # chemin vers le dossier des logos

    )
  
```

```python
k = 100 # 1 frame toutes les k lignes

duration = 60000 # durée de la vidéo en ms

spp = 10 # steps per period, 10 permet d'avoir une transition fluide, et des dépassements corrects

bars = 10 # top 6 et plus, affichage correct des logos, en dessous pb

title = "CLASSEMENT PERPETUEL TOP " + str(bars) + " CLUBS FRANCAIS" 

file_to_save = "test_" + "k" + str(k) + "_duration" + str(int(duration/1000)) + "s" + "_spp" + str(spp) + "_top" + str(bars)

print(file_to_save)
```
